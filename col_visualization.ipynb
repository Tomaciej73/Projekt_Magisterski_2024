```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('egzoplanety_pl.csv', encoding='utf-8')

sns.set_style('whitegrid')

# 1. Zależność masy planety od odległości od gwiazdy
plt.figure(figsize=(10, 6))
data1 = df[['Pol_Os_Orbity', 'Masa']].dropna()
sns.scatterplot(data=data1, x='Pol_Os_Orbity', y='Masa')
plt.xlabel('Pół Oś Orbity (au)')
plt.ylabel('Masa Planety ($M_{Jup}$)')
plt.title('Zależność masy planety od odległości od gwiazdy')
plt.show()
```

```python
# 2. Czy cięższe planety częściej występują przy masywniejszych gwiazdach?
plt.figure(figsize=(10, 6))
data2 = df[['Masa_Gwiazdy', 'Masa']].dropna()
sns.scatterplot(data=data2, x='Masa_Gwiazdy', y='Masa')
plt.xlabel('Masa Gwiazdy ($M_{Sun}$)')
plt.ylabel('Masa Planety ($M_{Jup}$)')
plt.title('Zależność masy planety od masy gwiazdy')
plt.show()
```

```python
# 3. Czy masywniejsze planety mają tendencję do dłuższych okresów orbitalnych?
plt.figure(figsize=(10, 6))
data3 = df[['Okres_Orbitalny', 'Masa']].dropna()
sns.scatterplot(data=data3, x='Okres_Orbitalny', y='Masa')
plt.xlabel('Okres Orbitalny (dni)')
plt.ylabel('Masa Planety ($M_{Jup}$)')
plt.title('Zależność masy planety od okresu orbitalnego')
plt.show()
```

```python
# 4. Zależność między czasem obiegu planety a odległością od gwiazdy (Prawa Keplera)
plt.figure(figsize=(10, 6))
data4 = df[['Pol_Os_Orbity', 'Okres_Orbitalny']].dropna()
sns.scatterplot(data=data4, x='Pol_Os_Orbity', y='Okres_Orbitalny')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Pół Oś Orbity (au)')
plt.ylabel('Okres Orbitalny (dni)')
plt.title('Zależność okresu orbitalnego od pół osi orbity (skala log-log)')
plt.show()
```

```python
# 5. Czy planety krążące wokół cieplejszych gwiazd mają tendencję do większych mas?
plt.figure(figsize=(10, 6))
data5 = df[['Efektywna_Temperatura_Gwiazdy', 'Masa']].dropna()
data5['Efektywna Temperatura Gwiazdy (°C)'] = data5['Efektywna_Temperatura_Gwiazdy'] - 273.15
sns.scatterplot(data=data5, x='Efektywna_Temperatura_Gwiazdy', y='Masa')
plt.xlabel('Efektywna Temperatura Gwiazdy (°C)')
plt.ylabel('Masa Planety ($M_{Jup}$)')
plt.title('Zależność masy planety od temperatury gwiazdy')
plt.show()
```

```python
# 6. Warunki na egzoplanecie - rozkład temperatury równowagi
plt.figure(figsize=(10, 6))
data6 = df['Temperatura_Rownowagi'].dropna() - 273.15
sns.histplot(data=data6, bins=30)
plt.ylabel('Liczba egzoplanet')
plt.xlabel('Temperatura Równowagi (°C)')
plt.title('Rozkład temperatury równowagi na egzoplanetach')
plt.show()
```
